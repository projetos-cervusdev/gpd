# 2. Interpretação do IQR (Intervalo Interquartílico)

 - **IQR:** O IQR indica a dispersão dos dados e a variabilidade dos resultados dentro da "faixa central" da distribuição (50% dos dados). Quanto maior o IQR, maior a variabilidade dos dados.

 - **Insight:** Se o IQR for muito grande, significa que existe uma alta variabilidade nas variações percentuais de crescimento das plantas ao longo do tempo. Isso pode sugerir que diferentes áreas ou condições do plantio não estão uniformemente controladas ou que diferentes fatores, como irrigação ou adubação, estão afetando o desenvolvimento das plantas de maneira desigual. Isso pode ser uma indicação de que o manejo da plantação precisa de ajustes.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Base de dados consolidada
data = {
    "Setor 1": {
        "DAP": [6, 7, 8, 9, 10, 11, 12, 13, 14, 18, 21, 27, 29, 32, 35, 39, 42, 45, 48, 52, 58, 80],
        "Variação (%)": [66.83, 74.47, 61.76, 55.16, 61.69, 63.84, 55.12, 51.43, 56.43, 49.53, 63.02, 
                         52.5, 33.35, 66.23, 15.18, 44.58, 12.69, 0, 75.86, 0, 0, 0]
    },
    "Setor 2": {
        "DAP": [6, 7, 8, 9, 10, 11, 12, 13, 14, 18, 21, 27, 29, 32, 35, 39, 42, 45, 48, 52, 58, 80],
        "Variação (%)": [58.73, 93.63, 72.18, 72.81, 63.4, 75.22, 78.5, 94.79, 85.89, 69.83, 47.04, 
                         43.68, 57.84, 57.24, 43.2, 69.99, 34.4, 69.7, 0, 0, 0, 0]
    },
    "Setor 3": {
        "DAP": [6, 7, 8, 9, 10, 11, 12, 13, 14, 18, 21, 27, 29, 32, 35, 39, 42, 45, 48, 52, 58, 80],
        "Variação (%)": [81.95, 66.04, 63.67, 44.68, 52.25, 62.33, 61.41, 61.35, 58.25, 84.77, 48.91, 
                         40.84, 38.56, 56.25, 40.04, 51.44, 59.86, 67.92, 0, 0, 0, 0]
    },
    "Setor 4": {
        "DAP": [6, 7, 8, 9, 10, 11, 12, 13, 14, 18, 21, 27, 29, 32, 35, 39, 42, 45, 48, 52, 58, 80],
        "Variação (%)": [0, 84.72, 86.17, 89.77, 71.67, 24.62, 89.86, 82.61, 80.23, 83.57, 77.52, 
                         48.96, 51.74, 56.27, 47.08, 44.83, 104.87, 72.4, 0, 0, 0, 0]
    },
    "Setor 5": {
        "DAP": [6, 7, 8, 9, 10, 11, 12, 13, 14, 18, 21, 27, 29, 32, 35, 39, 42, 45, 48, 52, 58, 80],
        "Variação (%)": [26.69, 75.64, 48.86, 44.52, 42.87, 60.28, 66.16, 63.06, 53.52, 62.25, 52.99, 
                         37.96, 1.08, 44.22, 0, 0, 35.29, 34.58, 5.96, 0, 0, 0]
    },
}

# Consolidando os dados
dap_values = []
cv_values = []

for setor_data in data.values():
    dap_values.extend(setor_data["DAP"])
    cv_values.extend(setor_data["Variação (%)"])

# Transformar listas em arrays
dap_values = np.array(dap_values).reshape(-1, 1)  # Para formato (n_samples, n_features)
cv_values = np.array(cv_values).reshape(-1, 1)

# Ajuste do modelo linear
linear_model = LinearRegression()
linear_model.fit(dap_values, cv_values)

# Coeficientes do modelo
slope = linear_model.coef_[0][0]
intercept = linear_model.intercept_[0]

# Predições do modelo
cv_pred = linear_model.predict(dap_values)

# Cálculo do Erro Quadrático Médio (MSE)
mse = mean_squared_error(cv_values, cv_pred)

# Exibição dos resultados
print(f"Modelo Linear (Geral): CV = {slope:.2f} * DAP + {intercept:.2f}")
print(f"MSE (Erro Quadrático Médio): {mse:.2f}")

# Visualização do modelo
plt.figure(figsize=(10, 6))
plt.scatter(dap_values, cv_values, color='blue', label='Valores Observados', alpha=0.6)
plt.plot(dap_values, cv_pred, color='red', label='Modelo Linear', linewidth=2)
plt.title("Regressão Linear (Todos os Setores): DAP vs CV")
plt.xlabel("DAP")
plt.ylabel("CV")
plt.legend()
plt.grid(True)
plt.show()
```
